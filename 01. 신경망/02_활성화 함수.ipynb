###활성화 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5, 5, 501)
```

- heaviside step function

```python
def heaviside(x):
    y = np.zeros_like(x)
    y[x > 0] = 1 # x <= 0 이면 0, x > 1 이면 1
    return y
```

```python
plt.plot(xs, heaviside(xs))
plt.grid()
plt.title("Heaviside step function")
plt.show()
```

- sigmoid


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.plot(xs, sigmoid(xs))
plt.grid()
plt.title("sigmoid")
plt.show()
```

- hyperbolic tangent

```python
plt.plot(xs, np.tanh(xs))
plt.grid()
plt.title("hyperbolic tangent")
plt.show()
```

- relu(rectified linear unit)

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(xs, relu(xs))
plt.grid()
plt.title("rectified linear unit")
plt.show()
```

- leaky relu

```python
def leaky_relu(x):
    return np.maximum(0.01 * x, x)
```

```python
plt.plot(xs, leaky_relu(xs))
plt.grid()
plt.title("leaky_relu")
plt.show()
```

- soft plus

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
plt.plot(xs, softplus(xs))
plt.grid()
plt.title("softplus")
plt.show()
```

- soft max

```python
def softmax(x):
    x = x - np.max(x)
    return np.exp(x) / np.sum(np.exp(x))
```

```python
yhat = np.array([-1, 1, 3])
print(softmax(yhat))
# 제일 큰 수를 1이라고 하고 나머지를 0 으로 두면 one-hot encoding을 쓸 수 있다
```

```python
np.sum(softmax(yhat))
```

```python
plt.plot(xs, softplus(xs))
plt.grid()
plt.title("softplus")
plt.show()
```
