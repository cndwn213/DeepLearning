# Cifar-10 이미지 분류 - v1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
plt.figure(figsize=(16,4))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(X_train[i]), plt.axis("off")
    plt.title(class_names[y_train[i,0]], fontsize=8)
plt.show
```

```python
# X_data : scaling
X_train = X_train/255.
X_test = X_test/255.
```

```python
# y_data : onehot encoding
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

- 모델 정의/설정/학습/평가

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential([
    Conv2D(32, (3,3), input_shape = X_train.shape[1:], padding="same", activation = "relu"),
    Conv2D(32, (3,3), activation = "relu"),
    MaxPooling2D(),
    Dropout(0.25),
    Conv2D(64, (3,3), padding = "same", activation = "relu"),
    Conv2D(32, (3,3), activation = "relu"),
    MaxPooling2D(),
    Dropout(0.25),
    Flatten(),
    Dense(512, activation = "relu"),
    Dropout(0.5),
    Dense(10, activation = "softmax"),
])
model.summary()
```

```python
model.compile(optimizer = RMSprop(learning_rate=0.0001, decay=1e-6),
              loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
model_path = "best-cifar10-v1.h5"
mc = ModelCheckpoint(model_path, save_best_only=True, verbose=1)
es = EarlyStopping(patience=10)
```

```python
hist = model.fit(X_train, y_train, validation_split=0.2,
                 epochs=100, batch_size=32, callbacks=[mc,es])
```

```python
y_acc = hist.history["accuracy"]
y_vloss = hist.history["val_loss"]
xs = np.arange(1, len(y_acc) + 1)
```

```python
plt.figure(figsize = (12,8))
plt.plot(xs, y_acc, label = "Accuracy")
plt.plot(xs, y_vloss, label = "Validation loss")
plt.legend(), plt.grid(), plt.xlabel("Epoch"), plt.title("MNIST CNN model v1"), plt.ylim([0,1])
plt.show()
```
